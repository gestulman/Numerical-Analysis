```python
import numpy as np
import pandas as pd

def f(t,y):
  return (2/t)*y + t**2 * np.exp(t)

#INPUT endpoints a, b; integer N; initial condition alpha
def euler(a, b, alpha, N):
  #set h, t and w
  h = (b-a)/N
  t = a
  w = alpha
  estimates = [(t,w)]
  for i in range(N):
    w = w + h * f(t,w) #compute w_i
    w = float(w)
    t = a + (i+1) * h #compute t_i
    estimates.append((t,w))
  return estimates

def y_exact(t):
  return t**2 * (np.exp(t) - np.exp(1))
```

```python
# Part (a) h = 0.1
estimates1 = euler(1,2,0,10)

chart = []
for t, w in estimates1:
    y_i = y_exact(t)
    error = abs(y_i - w)
    chart.append([t, w, y_i, error])

df = pd.DataFrame(chart, columns=["t_i", "w_i", "y_i", "|y_i - w_i|"])
df
```

```python

# Part (b): Linear interpolation for 1.04, 1.55, 1.97

# Euler with h = 0.1 → N = 10 for interval [1, 2]
estimates = euler(1, 2, 0, 10)

targets = [1.04, 1.55, 1.97]
results = []

for t_star in targets:
    # find the correct interval [t_i, t_{i+1}]
    for (t_i, w_i), (t_j, w_j) in zip(estimates[:-1], estimates[1:]):
        if t_i <= t_star <= t_j:
            # linear interpolation formula
            w_star = w_i + (t_star - t_i) / (t_j - t_i) * (w_j - w_i)
            y_true = y_exact(t_star)
            error = abs(y_true - w_star)
            results.append([t_star, w_star, y_true, error])
            break
df_partb = pd.DataFrame(results,
                        columns=["t", "Interpolated w(t)", "Exact y(t)", "Error"])
df_partb
```

```python
# Part (c) Step Size Calculation

tolerance = 0.1
a, b = 1, 2

# Determine Lipschitz Constant L (L = max(|df/dy|))
def df_dy(t):
    return 2.0 / t
t_vals = np.linspace(a, b, 100)
L = max(df_dy(t) for t in t_vals)

# Determine M (M = max(|y''(t)|))
def y_double_prime_exact(t):
    return abs(np.exp(t) * (t**2 + 4*t + 2) - 2 * np.exp(1))

# Estimate M by finding the maximum value on [a, b]
M = max(y_double_prime_exact(t) for t in np.linspace(a, b, 1000))

# Compute h using the Euler error bound formula (Eq. 5.10)
numerator = 2 * L * tolerance
denominator = M * (np.exp(L * (b - a)) - 1)
h = numerator / denominator
print("Step size h for tolerance 0.1:", h)
```
