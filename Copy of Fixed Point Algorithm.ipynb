```python
#INPUT: function g, initial approximation p_0, tolerance TOL, maximum iterations N
#OUTPUT: approximate solution p or failure
def fixed_point(g, p_0, TOL, N):
  #STEP 1: set i = 1
  i = 1

  #STEP 2: while i <= N do steps 3-6
  while i <= N:

    #STEP 3: set p = g(p_0)
    p = g(p_0) #compute p_i

    #STEP 4: if |p - p_0| < TOL, then output p
    if abs(p - p_0) < TOL:
      print(f'The procedure is successful with an approximate solution of {p} after {i} iteration(s).')
      return p #successful procedure

    #STEP 5: set i = i + 1
    i = i + 1

    #STEP 6: set p_0 = p
    p_0 = p #update p_0

  #STEP 7: message of failure
  print(f'The procdure failed after {N} iterations.')
  return #the procedure was unsuccessful
```

```python
#2.2 Problem 14a

import math
def show_iterations(g, p0, TOL=1e-5, N=100, show_first=6):
    p = p0
    print(f"p0 = {p0}")
    for n in range(1, show_first + 1):
        p = g(p)
        print(f"p{n} = {p}")

    p_final = fixed_point(g, p0, TOL, N)
    if p_final is not None:
        print(f"Final approximate solution: {p_final}\n")
    else:
        print("Did not converge within N iterations.\n")
    return p_final

# Problem 14(a) 2 + sin(x) - x = 0

def g14a(x): return 2 + math.sin(x)


print("Problem 14a sequence")
show_iterations(g14a, p0=2.5)
```

```python
#2.2 Problem 2

import math

# Define f(x)

def f(x):
    return x**4 + 2*x**2 - x - 3

# Define g1–g4

def g1(x): return (3 + x - 2*x**2) ** (1/4)
def g2(x): return math.sqrt((x + 3 - x**4) / 2)
def g3(x): return math.sqrt((x + 3) / (x**2 + 2))
def g4(x): return (3*x**4 + 2*x**2 + 3) / (4*x**3 + 4*x - 1)


g_functions = [g1, g2, g3, g4]

# Track best approximation
best_g = None
min_error = float('inf')

print("Problem 2a sequences")

for i, g in enumerate(g_functions, start=1):
    print(f"\nFunction g{i}:")
    p = 1  # initial guess
    print(f"p0 = {p}")

    # Perform 6 iterations
    for n in range(6):
        p = g(p)
        print(f"p{n+1} = {p}")

    # Run fixed-point iteration for final approximation
    approx = fixed_point(g, p_0=1, TOL=1e-5, N=30)
    if approx is not None:
        error = abs(f(approx))
        print(f"Final approx: {approx}, |f({approx})| = {error}")

        # Update best approximation
        if error < min_error:
            min_error = error
            best_g = f"g{i}"
    else:
        print("Did not converge within 30 iterations.")

print(f"\nBest approximation is given by {best_g} with |f(p)| = {min_error}")
```
