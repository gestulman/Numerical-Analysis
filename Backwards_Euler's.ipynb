```python
import numpy as np
import pandas as pd

def f(t,y):
  return -9*y

#INPUT endpoints a, b; integer N; initial condition alpha
def forward_euler(a, b, alpha, h):
  #set N, t and w
  N = int((b-a)/h)
  t = a
  w = alpha
  estimates = [(t,w)]
  for i in range(N):
    w = w + h * f(t,w) #compute w_i
    w = float(w)
    t = a + (i+1) * h #compute t_i
    estimates.append((t,w))
  return estimates

def backward_euler(a, b, alpha, h):
  #set N, t and w
  N = int((b-a)/h)
  t = a
  w = alpha
  estimates = [(t,w)]
  for i in range(N):
    w = w / (1+9*h) #compute w_i
    w = float(w)
    t = a + (i+1) * h #compute t_i
    estimates.append((t,w))
  return estimates

def y_exact(t):
  return np.exp(1-9*t)
```

```python
estimates1 = forward_euler(0,1,np.exp(1),0.1)

chart = []
for t, w in estimates1:
    y_i = y_exact(t)
    error = abs(y_i - w)
    chart.append([t, w, y_i, error])

df = pd.DataFrame(chart, columns=["t_i", "w_i", "y_i", "|y_i - w_i|"])
df
```

```python
def find_steps_for_accuracy():
    # Parameters
    a = 0.0
    b = 1.0
    # Initial condition: y(0) = e^(1 - 9*0) = e^1
    alpha = np.exp(1)

    # Calculate exact value at t=b for comparison
    true_value = y_exact(b) # e^-8 approx 0.000335

    tolerances = [0.1, 0.01, 0.001]

    print(f"{'Tolerance':<12} | {'Method':<15} | {'N':<5} | {'Final w':<12} | {'Error'}")
    print("-" * 65)

    for tol in tolerances:
        # 1. Find N for Forward Euler
        n_fe = 1
        while True:
            h = (b-a)/n_fe
            # Get the final estimate (last element of list, second item in tuple)
            fe_est = forward_euler(a, b, alpha, h)[-1][1]

            error = abs(true_value - fe_est)
            if error < tol:
                print(f"{tol:<12} | {'Forward Euler':<15} | {n_fe:<5} | {fe_est:.6f}     | {error:.6f}")
                break
            n_fe += 1
            if n_fe > 100000: break # Safety break

        # 2. Find N for Backward Euler
        n_be = 1
        while True:
            h = (b-a)/n_be
            be_est = backward_euler(a, b, alpha, h)[-1][1]
            error = abs(true_value - be_est)
            if error < tol:
                print(f"{tol:<12} | {'Backward Euler':<15} | {n_be:<5} | {be_est:.6f}     | {error:.6f}")
                break
            n_be += 1
            if n_be > 100000: break # Safety break

        print("-" * 65)

if __name__ == "__main__":
    find_steps_for_accuracy()
```
