```python
#newton's method
#import package to compute derivates
import sympy as sp

#INPUT: sympy expression for function f, initial approximation p_0, tolerance TOL, maximum iterations N
#OUTPUT: approximate solution p or failure
def newtons_method(f_expr, p_0, TOL, N):
  x = sp.Symbol('x')
  f = sp.lambdify(x, f_expr, "math")
  df = sp.lambdify(x, sp.diff(f_expr, x), "math")

  #STEP 1: set i = 1
  i = 1

  #STEP 2: while i <= N do steps 3-6
  while i <= N:

    #check if derivate is 0
    if df(p_0) == 0:
      print(f'Derivative at {p_0} is 0. The procedure has failed.')

    #STEP 3: set p = p_0 - f(p_0)/f'(p_0)
    p = p_0 - f(p_0)/df(p_0) #compute p_i

    #STEP 4: if |p - p_0| < TOL, then output p
    if abs(p - p_0) < TOL:
      print(f'The procedure is successful with an approximate solution of {p} after {i} iteration(s)')
      print(f'and tolerance of {TOL}.')
      return #successful procedure

    #STEP 5: set i = i + 1
    i = i + 1

    #STEP 6: set p_0 = p
    p_0 = p #update p_0

  #STEP 7: message of failure
  print(f'The procdure failed after {N} iterations.')
  return #the procedure was unsuccessful
```

```python
#secant method
#INPUT: function f, initial approximations p_0 and p_1, tolerance TOL, maximum iterations N
#OUTPUT: approximate solution p or failure
def secant_method(f, p_0, p_1, TOL, N):

  #STEP 1: set i = 2, q_0 = f(p_0), q_1=f(p_1)
  i = 2
  q_0 = f(p_0)
  q_1 = f(p_1)

  #STEP 2: while i <= N do steps 3-6
  while i <= N:

    if q_1 - q_0 == 0:   # avoid division by zero
      print(f'Method unsuccessful due to division by zero during iteration {i}.')
      return

    #STEP 3: set p = p_1 - q_1(p_1 - p_0)/(q_1 - q_0)
    p = p_1 - q_1*(p_1 - p_0)/(q_1 - q_0) #compute p_i

    #STEP 4: if |p - p_1| < TOL, then output p
    if abs(p - p_1) < TOL:
      print(f'The procedure is successful with an approximate solution of {p} after {i} iteration(s)')
      print(f'and tolerance of {TOL}.')
      return #successful procedure

    #STEP 5: set i=i+1
    i = i + 1

    #STEP 6: set p_0 = p_1, q_0 = q_1, p_1 = p, q_1 = f(p)
    p_0 = p_1 #update p_0
    q_0 = q_1 #update q_0
    p_1 = p #update p_1
    q_1 = f(p) #update q_1

  #STEP 7: message of failure
  print(f'The procdure failed after {N} iterations.')
  return #the procedure was unsuccessful
```
