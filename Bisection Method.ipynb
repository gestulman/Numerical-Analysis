```python
import numpy as np

#bisection method with:
#INPUT f = function, a,b = endpoints, TOL = tolerance, N = maximum number of iterations
#OUTPUT p = approximate solution
def bisection(f, a, b, TOL, N):

  #STEP 1
  i = 1
  FA = f(a)

  #print out starting interval
  print(f"On the starting interval of [{a},{b}]:")

  #STEP 2
  while i <= N:

    #STEP 3
    p = a + (b-a)/2 #compute p_i
    FP = f(p)

    #STEP 4
    if FP == 0 or (b-a)/2 < TOL:
      print(f"Method successful with an approximate solution of {p} and a tolerance of {TOL}.") #procedure completed successfully
      return

    #STEP 5
    i = i + 1

    #STEP 6
    if FA * FP > 0:
      a = p #compute a_i and b_i
      FA = FP
    else:
      b = p #f(a) is unchanged

  print(f"Method failed after {N} iterations.") #procedure unsuccessful
```

```python
#problem 6a
def f1(x):
  return 3 * x - np.exp(x)

bisection(f1, 1, 2, 1e-5, 50)
```

```python
#problem 6b
def f2(x):
  return 2 * x + 3 * np.cos(x) - np.exp(x)

bisection(f2, 0, 1, 1e-5, 50)
```

```python
#problem 6c
def f3(x):
  return x**2 - 4 * x + 4 - np.log(x)

bisection(f3, 1, 2, 1e-5, 50)
bisection(f3, 2, 4, 1e-5, 50)
```

```python
#problem 6d
def f4(x):
  return x + 1 - 2 * np.sin(np.pi * x)

bisection(f4, 0, 0.5, 1e-5, 50)
bisection(f4, 0.5, 1, 1e-5, 50)
```
