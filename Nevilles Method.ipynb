```python
import numpy as np
def Neville(x, xi, fi):
  n = len(xi)
  Q = np.zeros((n, n))
  for i in range(n):
    Q[i, 0] = fi[i]
  for i in range(1, n):
    for j in range(1, i + 1):
      Q[i, j] = ((x - xi[i - j]) * Q[i, j - 1] - (x - xi[i]) * Q[i - 1, j - 1]) / (xi[i] - xi[i - j])
  return Q
```

```python
xia = [0,0.25,0.5,0.75]
fia = [1,1.64872,2.71828,4.48169]
xa=0.43
Qa = Neville(xa,xia,fia)
```

```python
p1a = Qa[1,1]
p2a = Qa[2,2]
p3a = Qa[3,3]
print("degree 1 =",p1a,"degree 2 =",p2a,"degree 3 =",p3a)
```
