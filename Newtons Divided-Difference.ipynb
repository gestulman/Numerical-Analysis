```python
def Newtons(xi, fi):
  n = len(xi)
  F = [[0 for _ in range(n)] for _ in range(n)]
  for i in range(n):
    F[i][0] = fi[i]
  for j in range(1, n):
    for i in range(j, n):
      F[i][j] = (F[i][j-1] - F[i-1][j-1]) / (xi[i] - xi[i-j])
  return F
def Newtons_polynomial(x, xi, F):
  n = len(xi)
  result = F[0][0]
  product = 1
  for i in range(1, n):
    product *= (x - xi[i-1])
    result += F[i][i] * product
  return result
```

```python
xia = [-0.75,-0.5,-0.25,0]
fia = [-0.07181250,-0.02475000,0.33493750,1.10100000]
xa = -0.333333333
Fa = Newtons(xia,fia)
p1a = Newtons_polynomial(xa,xia[:2],[[Fa[i][j] for j in range(2)] for i in range(2)])
p2a = Newtons_polynomial(xa,xia[:3],[[Fa[i][j] for j in range(3)] for i in range(3)])
p3a = Newtons_polynomial(xa,xia[:4],[[Fa[i][j] for j in range(4)] for i in range(4)])
print("degree 1 =", p1a, "degree 2 =", p2a, "degree 3 =", p3a)
```
